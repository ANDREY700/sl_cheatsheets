```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(666)
x = np.arange(0, 10)
y = np.sin(x) + np.random.normal(0, .4, size=len(x))
```

```python
x_test = np.arange(9, 20)
y_test = np.sin(x_test) + np.random.normal(0, .2, size=len(x_test))
```

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 4))

ax.scatter(x, y, label='train')
ax.scatter(x_test, y_test, marker='*', label='test')
ax.set_ylim(-3, 3)



for i in range(1, 5):
    pf = PolynomialFeatures(i)
    x_ = pf.fit_transform(x.reshape(-1, 1))
    x_test_ = pf.transform(x_test.reshape(-1, 1))
    lr = LinearRegression()
    y2 = lr.fit(x_, y).predict(x_)
    y3 = lr.predict(x_test_)
    train_mse = mean_squared_error(y, y2)
    test_mse = mean_squared_error(y_test, y3)
    ax.plot(x, y2, c=colors[i], label=f'Degree: {i}, train_mse: {train_mse:.3f}, test_mse: {test_mse:.3f}')
    ax.plot(x_test, y3, c=colors[i])
    ax.legend(loc='upper right', bbox_to_anchor=(1., 1.5))
    # plt.show()
```

```python
ax.plot(1, 1)
```
